# **0. Описание задачи.**

Сердечные приступы или инфаркты миокарда продолжают оставаться серьезной глобальной проблемой здравоохранения, что требует более глубокого понимания их предшественников и потенциальных смягчающих факторов.

Набор данных для прогнозирования риска сердечного приступа

Атрибуты:


*   Patient ID - уникальный идентификатор для каждого пациента.
*   Age - Возраст пациента
*   Sex - Пол пациента (Мужчина/Женщина)
*   Cholesterol - Уровень холестерина у пациента
*   Blood Pressure - Артериальное давление пациента (систолическое/диастолическое)
*   Heart Rate - Частота сердечных сокращений пациента
*   Diabetes - есть ли у пациента диабет (да/нет)
*   Family History - Семейный анамнез проблем с сердцем (1: Да, 0: Нет)
*   Smoking - Курительный статус пациента (1: Курильщик, 0: Некурящий)
*   Obesity - статус ожирения пациента (1: ожирение, 0: отсутствие ожирения)
*   Alcohol Consumption - уровень употребления алкоголя пациентом (Нет/Легкий/Умеренный/Тяжелый)
*   Exercise Hours Per Week - количество часов тренировок в неделю.
*   Diet - Пищевые привычки пациента (Здоровый/Средний/Нездоровый)
*   Previous Heart Problems - предыдущие проблемы с сердцем у пациента (1: Да, 0: Нет)
*   Medication Use - использование лекарств пациентом (1: Да, 0: Нет)
*   Stress Level - уровень стресса, сообщаемый пациентом (1–10).
*   Sedentary Hours Per Day -часы сидячей активности в день
*   Income - Уровень дохода пациента
*   BMI - индекс массы тела (ИМТ) пациента
*   Triglycerides - уровень триглицеридов у пациента
*   Physical Activity Days Per Week - количество дней физической активности в неделю.
*   Sleep Hours Per Day - Часы сна в день
*   Country - Страна пациента
*   Continent - Континент, на котором проживает пациент.
*   Hemisphere - полушарие, в котором проживает пациент.
*   Heart Attack Risk - наличие риска сердечного приступа (1: Да, 0: Нет)

# **6. Разбиение данных на обучающую и тестовую выборки.**

```python
X = data[data.columns[:-1]]
y = data['Heart Attack Risk']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)

N_train, _ = X_train.shape
N_test,  _ = X_test.shape

N_train, N_test
```

```python
X_train
```

```python
X_test
```

# **7. Запустить классификатор (регрессию) ближайших соседей**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

k_n = KNeighborsClassifier(n_neighbors=3)
k_n.fit(X_train, y_train)

y_train_predict = k_n.predict(X_train)
y_test_predict = k_n.predict(X_test)

print('err_train=', np.mean(y_train != y_train_predict), 'err_test=', np.mean(y_test != y_test_predict))
print(f'Accuracy Score train : {accuracy_score(y_train, y_train_predict)*100}')
print(f'Accuracy Score test : {accuracy_score(y_test, y_test_predict)*100}')
```

Ошибка на тренировочной и тестовой выборках составила  23,3% и 43,8% соответственно. Попробуем подобрать оптимальный параметр для метода к-ближайших соседей.

# **8. Вычислить ошибки на обучающей и тестовой выборках. Выбрать оптимальное значение гиперпараметра (к-ва ближайших соседей). Сделать выводы.**

```python
from sklearn.model_selection import GridSearchCV
nnb = list(range(1, 30))
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)
best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

k_n = KNeighborsClassifier(n_neighbors=26)
k_n.fit(X_train, y_train)

y_train_predict = k_n.predict(X_train)
y_test_predict = k_n.predict(X_test)

print('err_train=', np.mean(y_train != y_train_predict), 'err_test=', np.mean(y_test != y_test_predict))
print(f'Accuracy Score train : {accuracy_score(y_train, y_train_predict)*100}')
print(f'Accuracy Score test : {accuracy_score(y_test, y_test_predict)*100}')
```

Ошибка на тренировочной и тестовой выборках составила 35,8% и 37,1% соответственно.

```python
from sklearn.svm import SVC
svc = SVC(gamma='auto')
svc.fit(X_train, y_train)

print('err_train=', np.mean(y_train != svc.predict(X_train)), 'err_test=', np.mean(y_test  != svc.predict(X_test)))
print(f'Accuracy Score train : {accuracy_score(y_train, y_train_predict)*100}')
print(f'Accuracy Score test : {accuracy_score(y_test, y_test_predict)*100}')
```

Ошибка на тренировочной и тестовой выборках составила 3,7% и 35,5% соответственно.

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 1000)
rf.fit(X_train, y_train)

print('err_train=', np.mean(y_train != rf.predict(X_train)), 'err_test=', np.mean(y_test  != rf.predict(X_test)))
print(f'Accuracy Score train : {accuracy_score(y_train, y_train_predict)*100}')
print(f'Accuracy Score test : {accuracy_score(y_test, y_test_predict)*100}')
```

Ошибка на тренировочной и тестовой выборках составила 0% и 0% соответственно.
